```python
import pandas as pd

# Load the dataset
# Ensure 'agricultural_yield_dataset.csv' is uploaded to your Colab session
df = pd.read_csv('agriculture_yield_dataset.csv')

# Q1. Dataset Overview
print("--- Q1: Dataset Overview ---")
print(f"Rows and Columns: {df.shape}") # [1]
print(f"Column Names: {df.columns.tolist()}") # [1]
print("First 10 Records:")
print(df.head(10)) # [1]

# Q2. Data Types and Missing Values
print("\n--- Q2: Data Types and Missing Values ---")
print(df.info()) # [1]
print("\nMissing Values per Column:")
print(df.isnull().sum()) # [1]

# Q3. Descriptive Statistics
print("\n--- Q3: Descriptive Statistics ---")
stats = df.describe() # [2]
print(stats)
# Identifying specific stats requested
print(f"\nFeature with highest mean: {stats.loc['mean'].idxmax()}") # [2]
print(f"Feature with highest standard deviation: {stats.loc['std'].idxmax()}") # [2]
```
