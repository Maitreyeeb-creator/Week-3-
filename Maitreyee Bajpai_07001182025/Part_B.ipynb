```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('agriculture_yield_dataset.csv')

# Q4 & Q7. Distribution Analysis
# Creating histograms for specific columns and checking for normality/outliers
cols_to_plot = ['rainfall_mm', 'temperature_c', 'fertilizer_kg', 'yield_ton_per_hectare'] # [2, 3]
df[cols_to_plot].hist(bins=20, figsize=(12, 10))
plt.suptitle("Feature Distributions and Yield Normality")
plt.show()

# Q5 & Q6. Crop and Soil Type Analysis
for col in ['crop_type', 'soil_type']: # [2, 3]
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col)
    plt.title(f"Frequency of {col}")
    plt.show()
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts()) # [2, 3]
```
