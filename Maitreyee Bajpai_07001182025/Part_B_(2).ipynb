```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('agriculture_yield_dataset.csv')

# Q8. Scatter Plot Analysis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
sns.scatterplot(data=df, x='rainfall_mm', y='yield_ton_per_hectare', ax=ax1) # [3]
ax1.set_title("Rainfall vs Yield")
sns.scatterplot(data=df, x='fertilizer_kg', y='yield_ton_per_hectare', ax=ax2) # [3]
ax2.set_title("Fertilizer vs Yield")
plt.show()

# Q9. Correlation Analysis
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['number']) # [4]
correlation_matrix = numeric_df.corr() # [4]
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') # [4]
plt.title("Correlation Heatmap")
plt.show()
print("\nTop 3 features correlated with yield:")
print(correlation_matrix['yield_ton_per_hectare'].sort_values(ascending=False)[1:4]) # [4]

# Q10. Group-Based Analysis
print("\nAverage yield by Crop Type:")
print(df.groupby('crop_type')['yield_ton_per_hectare'].mean()) # [4]
print("\nAverage yield by Soil Type:")
print(df.groupby('soil_type')['yield_ton_per_hectare'].mean()) # [4]
```
